```python
import os
import matplotlib.pyplot as plt
```

```python
base_folder = os.getcwd()
data_folder = os.path.join(base_folder, 'Dataset')
drowniness_folder = os.path.join(data_folder, 'Train')
```

```python
# Store counts
folder_names = []
image_counts = []

# Loop through class folders
for folder in os.listdir(drowniness_folder):
    folder_path = os.path.join(drowniness_folder, folder)
    
    if os.path.isdir(folder_path):
        images = [
            img for img in os.listdir(folder_path)
            if img.lower().endswith(('.png', '.jpg', '.jpeg'))
        ]
        folder_names.append(folder)
        image_counts.append(len(images))

# Plot bar chart
plt.figure()
plt.bar(folder_names, image_counts)
plt.xlabel("Class Folder")
plt.ylabel("Number of Images")
plt.title("Distribution of Images per Class")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
